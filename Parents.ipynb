```python
# create data object and create dataframe finaldf
data = psid.psiddata(rawdata = rawdata)
raw_finaldf = data.outputdf(varobj, threshold = 10)
```

```python
# replace missing values
nadict = {'GAS FOR HOME': {9999999: np.nan},
          'ELECTRICITY': {9999999: np.nan},
          'WATER/SEWER': {9999999: np.nan},
          'OTHER UTILITY': {9999999: np.nan}
}
raw_finaldf = raw_finaldf.replace(nadict)
head = raw_finaldf['relation_to_head'] == 10
spouse = raw_finaldf['relation_to_head'].isin([20,22])
# raw_finaldf.to_csv('new_psid.csv')
```

```python
conslist = ['FOOD', 'HOUSING', 'TRANSPORTATION', 'EDUCATION', 'CHILDCARE', 'HEALTH CARE']
raw_finaldf['total_cons'] = raw_finaldf[conslist].sum(axis = 1)
```

```python
# adjust for inflation using PCE price index
pce = pd.read_csv('PCEPI.csv')
# extract year
pce['year'] = pce['DATE'].str[:4]
pce['year'] = pd.to_numeric(pce['year'])
# normalize index to 1
pce['idx'] = pce['PCEPI'] / 100
pce = pce[['idx', 'year']]
finaldf = raw_finaldf.merge(pce, on = 'year')
finaldf['total_cons1'] = finaldf['total_cons'] / finaldf['idx']
finaldf['total_income'] = finaldf[['head_income', 'spouse_income']].sum(axis = 1)
finaldf['total_income1'] = finaldf['total_income'] / finaldf['idx']
finaldf['cons_over_inc'] = np.where(finaldf['total_income'] > 1000,
                                    finaldf['total_cons'] / finaldf['total_income'],
                                    np.nan)
```

```python
head = finaldf[finaldf['relation_to_head'] == 10]
parent = head[head['isparent'] == 1]
nonparent = head[head['isparent'] == 0]
```

```python
def graph_adjusted(df, var):
    fulldf = copy.deepcopy(df)
    fulldf['adj_var'] = df[var] / df['idx']
    graphdf = fulldf.groupby('age')['adj_var'].mean()[18:80]
    f = figure()
    f.line(x = graphdf.index, y = graphdf, legend = var)
    show(f)
```

```python
def compare_parent_nonparent(df, var, inf_adjusted = True):
    fulldf = copy.deepcopy(df)
    assert type(inf_adjusted) == bool
    if inf_adjusted == True:
        fulldf['adj_var'] = df[var] / df['idx']
    else:
        fulldf['adj_var'] = df[var]
    parent = fulldf['isparent'] == 1
    nonparent = np.invert(parent)
    parent_graphdf = fulldf[parent].groupby('age')['adj_var'].mean()[18:80]
    nonparent_graphdf = fulldf[nonparent].groupby('age')['adj_var'].mean()[18:80]
    f = figure(title = var)
    f.line(x = parent_graphdf.index, y = parent_graphdf, 
           legend = "parent")
    f.line(x = nonparent_graphdf.index, y = nonparent_graphdf, 
           legend = "nonparent", color = "red")
    show(f)
```

```python
parent_total_cons = parent.groupby('age')['total_cons1'].sum()
parent_total_inc = parent.groupby('age')['total_income1'].sum()
nonparent_total_cons = nonparent.groupby('age')['total_cons1'].sum()
nonparent_total_inc = nonparent.groupby('age')['total_income1'].sum()
tograph = (parent_total_cons / parent_total_inc)[18:80]
tograph1 = (nonparent_total_cons / nonparent_total_inc)[18:80]
f = figure()
f.line(x= tograph.index, y = tograph, legend = 'parent')
f.line(x = tograph1.index, y = tograph1, color = 'red', legend = 'non-parent')
f.legend.location = 'top_left'
show(f)
```
